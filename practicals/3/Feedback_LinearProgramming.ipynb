### **Example 1** : taken from Exercise1 of Session 5-2 (on pages 9-10)


```python
# Firstly, import the whole module of 'GEKKO' into your working environment, from the 'gekko' package
from gekko import GEKKO

# Secondly, create an instance 'm' of GEKKO problem, by calling the constructor method 'GEKKO()'
# Now problem components can be added to 'm', for instances, variables objective function, constraint functions
m = GEKKO()

# variables x (ice milk) and y (ice cream)
# Constraints on domains
x = m.Var(lb=0, ub=40) # domain of x is between 0 to 40
y = m.Var(lb=0, ub=40) # domain of y is between 0 to 40

# Objective function (maximise the Profit, P)
m.Maximize(8*x+10*y)

# Constraint functions
m.Equation(0.4*x+0.2*y<=10) # 10 litres of milk available
m.Equation(0.2*x+0.4*y<=14) # 14 litres of cream available

# Solving the linear programming problem 'm'
# You can turn on the diaply mode ('disp = True') to see log of solving process
m.solve(disp=False)

# Present the optimal values of x, y and the maximum profit
print ('Product 1 (x): ', round(x.value[0],1))
print ('Product 2 (y): ', round(y.value[0],1))
print ('Maximum profit is:', round(8*x.value[0] + 10*y.value[0],1))
```

### **Example 2** : description of a new case study is given below.


Assume a company is ordered to make exactly 40 gallons of mixture in which basic ingredients are x and y. With the aim to get the production cost as low as possible, **what should be amounts of x and y used?**


*   As for the cost, x costs £8 per gallon, while y costs £15 per gallon.
*   In the final mixture, there can be upto 12 gallons of x, and at least 10 gallons of y. 






```python
# Firstly, import the whole module of 'GEKKO' into your working environment, from the 'gekko' package
from gekko import GEKKO

# Secondly, create an instance 'm' of GEKKO problem, by calling the constructor method 'GEKKO()'
# Now problem components can be added to 'm', for instances, variables objective function, constraint functions
m = GEKKO()

# variables x and y
# Constraints on domains
x = m.Var(lb=0, ub=40) # domain of x is between 0 to 40
y = m.Var(lb=0, ub=40) # domain of y is between 0 to 40

# Objective function (minimise the cost)
m.Minimize(8*x+15*y)

# Constraint functions
m.Equation(x+y==40) # 40 litres of mixture is required
m.Equation(x<=12) # Up to 12 gallons of x
m.Equation(y>=10) # At least 10 gallons of y

# Solving the linear programming problem 'm'
# You can turn on the diaply mode ('disp = True') to see log of solving process
m.solve(disp=False)

# Present the optimal values of x, y and the minimum cost
print ('Ingredient 1 (x): ', round(x.value[0],1))
print ('Ingredient 2 (y): ', round(y.value[0],1))
print ('Minimum cost is:', round(8*x.value[0] + 15*y.value[0],1))
```

### **Question 1** :

Your firm makes two products (X and Y) using two machines (A and B). 


*   Each unit of X that is produced requires 50 minutes processing time on machine A, and 30 minutes processing time on machine B.
*   Each unit of Y that is produced requires 24 minutes processing time on machine A, and 33 minutes processing time on machine B.
*   At the start of the current week there are 30 units of X and 90 units of Y in stock. Available processing time on machine A is forecast to be 40 hours and on machine B is forecast to be 35 hours.
*   The demand for X in the current week is forecast to be 75 units and for Y is forecast to be 95 units.

The first aims is to maximise the combined sum of the units of X and the units of Y in stock at the end of the week.


*   *Expected result of objective value is 1.2*



```python
# Firstly, import the whole module of 'GEKKO' into your working environment, from the 'gekko' package
from gekko import GEKKO

# Secondly, create an instance 'm' of GEKKO problem, by calling the constructor method 'GEKKO()'
# Now problem components can be added to 'm', for instances, variables objective function, constraint functions
m = GEKKO()

# variables x (new units of product X) and y (new units of product Y)
# Constraints on their value domains
x = m.Var(lb=0, ub=100) # domain of x is between 0 to 100
y = m.Var(lb=0, ub=100) # domain of y is between 0 to 100

# Objective function (maximise the combined stock of X and Y)
# This is to maximise (x+30-75) + (y+90-95) = (x+y-50)
m.Maximize(x+y-50)

# Constraint functions
m.Equation(50*x + 24*y <= 40*60) # machine A time
m.Equation(30*x + 33*y <= 35*60) # machine B time
m.Equation(x>=45) # minimum of new units of X
m.Equation(y>=5) # minumum of new units of Y

# Solving the linear programming problem 'm'
# You can turn on the diaply mode ('disp = True') to see log of solving process
m.solve(disp=False)

# Present the optimal values of x, y and the maximum profit
print ('New units of X (x): ', round(x.value[0],1))
print ('New units of Y (y): ', round(y.value[0],1))
print ('Maximum combined stock of X and Y:', round(x.value[0] + y.value[0] - 50, 1))
```

### **Question 2** :

At Aberystwyth U., Prof T is to employ two people, Richi and Lizz, to grade papers from his module. Each must be employed at least one hour a week to justify their employment.


*   Richi is a graduate student and can grade 20 papers per hour. For that, he earns £15 per hour.
*   Lizz is a post-doctoral associate and can grade 30 papers per hour. She can earn £25 per hour.
*   There are at least 110 papers to be graded each week. 

How many hours per week should Prof T employ each person to minimize the cost?

*   *Expected result of objective value is 85.0*


```python
# Firstly, import the whole module of 'GEKKO' into your working environment, from the 'gekko' package
from gekko import GEKKO

# Secondly, create an instance 'm' of GEKKO problem, by calling the constructor method 'GEKKO()'
# Now problem components can be added to 'm', for instances, variables objective function, constraint functions
m = GEKKO()

# variables x (hours for Richi) and y (hours for Lizz)
# Constraints on their value domains
x = m.Var(lb=0, ub=10) # domain of x is between 0 to 10
y = m.Var(lb=0, ub=10) # domain of y is between 0 to 10

# Objective function (minimise the cost)
m.Minimize(15*x + 25*y)

# Constraint functions
m.Equation(20*x + 30*y >= 110) # combined effort must be at least cover 100 papers
m.Equation(x>=1) # minimum hour per week
m.Equation(y>=1) # minumum hour per week

# Solving the linear programming problem 'm'
# You can turn on the diaply mode ('disp = True') to see log of solving process
m.solve(disp=False)

# Present the optimal values of x, y and the maximum profit
print ('Hours for Richi (x): ', round(x.value[0],1))
print ('Hours for Lizz (y): ', round(y.value[0],1))
print ('Minimum cost:', round(15*x.value[0] + 25*y.value[0],1))
```
