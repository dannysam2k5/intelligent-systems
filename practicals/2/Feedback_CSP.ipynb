### **Example 2** : assigning values to two variables, with a constraint.

```python
# import classes/methods defined in this 'constraint' package into the working space
from constraint import *

# create an instance of CSP problem
problem = Problem()

# define variables and their value domains, then add them to be part of the problem,
# there are two variables 'a' and 'b',
# 'a' can take a value from a list of [1, 2, 3], and 
# 'b' can take a value from a list of [4, 5, 6]
problem.addVariable("a", [1,2,3])
problem.addVariable("b", [4,5,6])

# define a constraint 'a = 2b' as a function of variables 'x' and 'y'
# it returns 'True' only when this constraint is satisfied!
def constraint1(x, y):
  return 2*x == y

# add this constraint to the problem, as 'a' and 'b' bind with 'x' and 'y' of the function 'constraint1'
problem.addConstraint(constraint1, ["a", "b"])

# generate all possible solution(s) for the problem
solution = problem.getSolutions()

# report the number of solution(s)
print("There are", len(solution), "solution(s).")

# print them out !
print(solution)
```

### **Example 3** : assigning values to two variables, with two or more constraints.

```python
# Example 3: assigning values to two variables, with two or more constraints !!
# ----------------------------------------------------------------------------------------

# import classes/methods defined in this 'constraint' package into the working space
from constraint import *

# create an instance of CSP problem
problem = Problem()

# define variables and their value domains, then add them to be part of the problem,
# there are two variables 'a' and 'b',
# 'a' can take a value from a list of [1, 2, 3], and 
# 'b' can take a value from a list of [4, 5, 6]
problem.addVariable("a", [1,2,3])
problem.addVariable("b", [4,5,6])

# define a constraint 'a = 2b' as a function of variables 'x' and 'y'
# it returns 'True' only when this constraint is satisfied!
def constraint1(x, y):
  return 2*x == y

# define a constraint 'a%2 > 0' as a function of variable 'x'
# it returns 'True' only when this constraint is satisfied, i.e., 'a' must be an odd number!
def constraint2(x):
  return x%2 > 0

# add this constraint to the problem, as 'a' and 'b' bind with 'x' and 'y' of the function 'constraint1'
problem.addConstraint(constraint1, ["a", "b"])

# add this constraint to the problem, as 'a' binds with 'x' of the function 'constraint2'
problem.addConstraint(constraint2, ["a"])

# generate all possible solution(s) for the problem
solution = problem.getSolutions()

# report the number of solution(s)
if len(solution) > 1:
  print("There are", len(solution), "solution(s).")
else: print("There is", len(solution), "solution.")

# print them out !
print(solution)
```

### **Example 4** : declaring multiple variables with the same value domain, and using a **built-in** constraint function.

```python
# import classes/methods defined in this 'constraint' package into the working space
from constraint import *

# create an instance of CSP problem
problem = Problem()

# define variables and their value domains, then add them to be part of the problem,
# there are two variables 'a' and 'b',
# they can take any value from the list [1,2,3,4]
problem.addVariables(["a","b"], [1,2,3,4])

# add the built-in constraint 'AllDifferentConstraint()' to this problem
# this constraint does not allow variables to have the same value
problem.addConstraint(AllDifferentConstraint())

# generate all possible solution(s) for the problem
solution = problem.getSolutions()

# report the number of solution(s)
if len(solution) > 1:
  print("There are", len(solution), "solution(s).")
else: print("There is", len(solution), "solution.")

# print them out !
print(solution)
```
